#Media Truncada
La usamos para sacar una media nueva que no involucre outliers.
###¿Cómo se hace?
Tenemos que decidir más o menos cuánto queremos quitar de los datos y tenemos que asegurar que sea lo mismo de un extremo que del otro.
No se debe de quitar demasiado, un 5% podría estar bien (quitamos 2.5% de cada extremo).


> Esto solo nos lo podemos permitir si tenemos muuuuchos datos. En dado caso de que sean pocos es preferible usar la mediana.



```python
stats.trim_mean(df["price"],0.1)
```

#Medidas de dispersión


```python
#Buscamos las medidas como la standard deviation.
#Aquí le intentaré con numpy y con pandas
display(df['price'].std())
np.std(df['price'])
```

```python
#Vamos a sacar los valores necesarios para las otras medidas de dispersión
price_max=df['price'].max()
price_min=df['price'].min()
print(price_max)
print(price_min)
rango=price_max-price_min
print(f'El Rango de mi data set es: {rango}')
```
